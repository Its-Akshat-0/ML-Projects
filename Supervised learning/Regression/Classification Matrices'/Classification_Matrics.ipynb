```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python

clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python

clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python

y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python

confusion_matrix(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred2) 
```

```python
print("precision of lr",precision_score(y_test,y_pred1,average=None))
print("recall of lr",recall_score(y_test,y_pred1))
print("f1_score of lr",f1_score(y_test,y_pred1))
```

```python
# for decission tree
print("precision of dt",precision_score(y_test,y_pred2))
print("recall of dt",recall_score(y_test,y_pred2))
print("f1_score of dt",f1_score(y_test,y_pred2))
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

# As a dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python

clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python

clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python

y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred2) 
```

```python
print("precision of lr",precision_score(y_test,y_pred1,average='micro'))
print("recall of lr",recall_score(y_test,y_pred1))
print("f1_score of lr",f1_score(y_test,y_pred1))
```

```python
# for decission tree
print("precision of dt",precision_score(y_test,y_pred2))
print("recall of dt",recall_score(y_test,y_pred2))
print("f1_score of dt",f1_score(y_test,y_pred2))
```
