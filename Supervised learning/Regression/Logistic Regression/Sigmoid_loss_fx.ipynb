```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
class SigmoidPerc:
    def __init__(self):
        self.intercept=None
        self.coef=None

    def sigmoid(self,y):
        return (1/(1+np.exp(-y)))
    def fit(self,x,y):
        lr=0.1
        x= np.insert(x,0,1,axis=1)
        weights = np.ones(x.shape[1])
        for i in range(1000):
            j=np.random.randint(0,x.shape[0])
            y_hat=self.sigmoid(np.dot(x[j],weights))
            weights=weights + lr * (y[j]-y_hat) * x[j]
        return weights[0],weights[1:]
         
```

```python
myperceptron=SigmoidPerc()
```

```python
intercept_,coef_ = myperceptron.fit(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=50)
plt.ylim(-3,2)
```

```python
class myPerceptron:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        
    def step(self,z):
        return 1 if z>0 else 0
    def fit(self,X,Y):
        X = np.insert(X,0,1,axis=1)
        weights = np.ones(X.shape[1])
        lr = 0.1
    
        for i in range(1000):
            j = np.random.randint(0,X.shape[0])
            y_hat = self.step(np.dot(X[j],weights))
            weights = weights + lr*(Y[j]-y_hat)*X[j]
        self.coef_=weights[1:]
        self.intercept_=weights[0]
        return weights[0],weights[1:]

    def predict(self,X):
        return self.step( self.intercept_ +np.dot(X,self.coef_))
```

```python
Percep_trick=myPerceptron()
intercept_,coef_ = Percep_trick.fit(X,y)
```
